```
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Nome': ['Alice', 'Joao', 'Charlie', 'David', 'Eva', 'Diego', 'Denize', 'Claudio'],
    'Idade': [25, 30, 35, 40, 45, 60, 22, 24],
    'Profissão': ['Engenheiro', 'Médico', 'Professor', 'Advogado', 'Médico', 'Engenheiro', 'Mecanico', 'ADM'],
    'Salário': ['4500', '8000', '5000', '10000', '12000', '15000', '1200', '1500'],
    'Limite_Credito': ['2500', '4000', '4000', '1000', '10000', '2000', '500', '250'],
    'Historico_Inadimplencia': ['0', '0', '0', '1', '0', '1', '0', '1'],
    'Estado_Civil': ['Casamento', 'Casamento', 'Solteiro', 'Solteiro', 'Casamento', 'Solteiro', 'Solteiro', 'Solteiro'],
    'Imovel_Proprio': ['0', '0', '0', '1', '1', '1', '0', '0']
}
df = pd.DataFrame(data)

df['Salário'] = pd.to_numeric(df['Salário'], errors='coerce')
df['Limite_Credito'] = pd.to_numeric(df['Limite_Credito'], errors='coerce')

salario_por_nome = df.groupby('Nome')['Salário'].mean()
limite_credito_por_nome = df.groupby('Nome')['Limite_Credito'].mean()

x = range(len(salario_por_nome))
largura= 0.4

plt.figure(figsize=(12, 6))
plt.bar([pos - largura / 2 for pos in x], salario_por_nome, width=largura,color='royalblue', label='salário')

plt.bar([pos + largura / 2 for pos in x], limite_credito_por_nome, width=largura,color='firebrick', label='limite de Crédito')

for i, valor in enumerate(salario_por_nome):
  plt.text(i - largura / 2, valor, f"{valor:.2f}", ha='center', va='bottom', fontsize=9)

  for i, valor in enumerate(limite_credito_por_nome):
    plt.text(i + largura / 2, valor, f"{valor:.2f}", ha='center', va='bottom', fontsize=9)


plt.title('Salário e Limite de Crédito por Nome', fontsize=16)
plt.xlabel('nome', fontsize=12)
plt.ylabel('valores (R$)', fontsize=12)
plt.xticks(x, salario_por_nome.index, rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()




```
